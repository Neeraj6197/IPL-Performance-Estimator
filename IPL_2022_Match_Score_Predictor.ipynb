```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
matches = pd.read_csv('/content/IPL_Matches_2008_2022.csv')
matches.head()
```

```python
matches.info()
```

```python
#working on column 'city'
matches[matches['City'].isnull()]
```

```python
for i in range(len(matches)):
    if (matches['Venue'] == 'Dubai International Cricket Stadium')[i] == True:
        matches['City'][i] = 'Dubai'
    elif (matches['Venue'] == 'Sharjah Cricket Stadium')[i] == True:
        matches['City'][i] = 'Sharjah'
    else:
        matches['City'][i] = matches['City'][i]
```

```python
matches['City'].isnull().sum()
```

```python
# checking filled values:
matches[(matches['ID'] == 1216495) | (matches['ID'] == 1216530)]
```

```python
#importing deliveries data:
delivery = pd.read_csv('/content/IPL_Ball_by_Ball_2008_2022.csv',error_bad_lines=False)
delivery.head()
```

```python
# calculating total runs after each inning:
temp_df = delivery.groupby(['ID','innings'])['total_run'].sum().reset_index()
```

```python
temp_df
```

```python
#we only need 1st inning score for further calculations:
temp_df = temp_df[temp_df['innings'] == 1]
```

```python
#merging temp_df with matches df:
match_df = matches.merge(temp_df,on='ID')
```

```python
match_df.head()
```

```python
#working on teams:
match_df['Team1'].unique()
```

```python
#re-nameing same teams:

#'Delhi Daredevils' --> 'Delhi Capitals':
match_df['Team1'] = match_df['Team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['Team2'] = match_df['Team2'].str.replace('Delhi Daredevils','Delhi Capitals')

#'Deccan Chargers' --> 'Sunrisers Hyderabad':
match_df['Team1'] = match_df['Team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['Team2'] = match_df['Team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

#'Gujarat Lions' --> 'Gujarat Titans'
match_df['Team1'] = match_df['Team1'].str.replace('Gujarat Lions','Gujarat Titans')
match_df['Team2'] = match_df['Team2'].str.replace('Gujarat Lions','Gujarat Titans')

```

```python
#keeping only those teams who played till 2022:
teams = ['Royal Challengers Bangalore', 'Kings XI Punjab',
       'Delhi Capitals', 'Mumbai Indians', 'Kolkata Knight Riders',
       'Rajasthan Royals','Chennai Super Kings','Sunrisers Hyderabad','Gujarat Titans'
        ,'Lucknow Super Giants']

match_df = match_df[match_df['Team1'].isin(teams)]
match_df = match_df[match_df['Team2'].isin(teams)]
```

```python
match_df['Team1'].unique()
```

'''
required cols for our final_df to predict the result would be:

('batting_team',
'bowling_team',
'city',
'current_score',
'overs_completed',
'wickets_left',
'crr',
'runs_in_last_3,
'total_runs',
)
'''

```python
match_df = match_df[['ID','City','Team1','Team2','total_run']]
```

```python
match_df.head()
```

```python
#merging match_df with delivery to get ball by ball details:
delivery_df = delivery.merge(match_df,on='ID')
```

```python
#keeping values where inning = 1 to predict scored for 1st innings:
delivery_df = delivery_df[delivery_df['innings'] == 1]
```

```python
delivery_df.head()
```

```python
#calculating the current score after every ball:
delivery_df['current_score'] = delivery_df.groupby('ID')['total_run_x'].cumsum()
```

```python
#calculating wickets left after every ball:
delivery_df['wickets_left'] = 10 - (delivery_df.groupby('ID')['isWicketDelivery'].cumsum())
```

```python
#calculating balls left after every ball:
delivery_df['balls_left'] = 120 - (delivery_df['overs']*6 + delivery_df['ballnumber'])
```

```python
delivery_df
```

```python
#calculating current run rate(crr):
delivery_df['crr'] = (delivery_df['current_score']*6)/(120-delivery_df['balls_left'])
```

```python
#creating bowling team as there is no bowling team column in the df:
```

```python
def bowling_team(row):
    return row['Team1'] if row['Team2'] == row['BattingTeam'] else row['Team2']
```

```python
delivery_df['BowlingTeam'] = delivery_df.apply(bowling_team,axis=1)
```

```python
#checking the bowling column with other cols:
delivery_df[['Team1','Team2','BattingTeam','BowlingTeam']]
```

```python
#calculating runs scored in last 3 overs:
groups = delivery_df.groupby('ID')
ids = delivery_df['ID'].unique()
last_three = []

for id in ids:
    last_three.extend(groups.get_group(id).rolling(window=18).sum()['total_run_x'].values)
```

```python
delivery_df['last_three'] = last_three
```

```python
delivery_df
```

```python
delivery_df.info()
```

```python
# gathering the required cols for final df:
final_df = delivery_df[['BattingTeam', 'BowlingTeam', 'City', 'current_score', 'overs', 'wickets_left', 'crr','last_three', 'total_run_y']]
```

```python
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
#removing 1st 3 overs data of every match which are NaN in our case:
final_df.dropna(inplace=True)
```

```python
#shuffling the df as we have sorted data:
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.shape
```

```python
final_df.head()
```

## Building the models:

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_df.iloc[:,:-1]
Y = final_df.iloc[:,-1]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train.shape
```

```python
Y_train.shape
```

## Creating pipelines to tranform cols and train the model:

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
# encoding batting team,bowling team and city:
step1 = ColumnTransformer(transformers=[
    ('ohe',OneHotEncoder(sparse=False,drop='first'),[0,1,2])
],remainder = 'passthrough')
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
#assigning our model to step 2 :
step2 = LinearRegression()
```

```python
#creating the pipeline:
pipe1 = Pipeline([
    ('step1',step1),
    ('step2',step2),
])
```

```python
#training the data with our pipeline:
pipe1.fit(X_train,Y_train)
```

```python
#testing the pipeline:
Y_pred1 = pipe1.predict(X_test)
```

```python
# evaluating the model:
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error,r2_score
```

```python
print('cvs:',cross_val_score(pipe1,X_train,Y_train,cv=5,scoring='r2').mean())
print('rmse:',mean_squared_error(Y_test,Y_pred1))
```

## Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
step2 = DecisionTreeRegressor(max_depth= 250,min_samples_split=2)
```

```python
#creating the pipeline:
pipe2 = Pipeline([
    ('step1',step1),
    ('step2',step2),
])
```

```python
#training the data with our pipeline:
pipe2.fit(X_train,Y_train)
```

```python
#testing the pipeline:
Y_pred2 = pipe2.predict(X_test)
```

```python
print('cvs:',cross_val_score(pipe2,X_train,Y_train,cv=5,scoring='r2').mean())
print('rmse:',mean_squared_error(Y_test,Y_pred2))
```

```python
#using randomized search cv for hyper parameter tuning:
from sklearn.model_selection import RandomizedSearchCV
```

```python
parameter = {
    'step2__max_depth' : [100,150,200,250,300],
    'step2__min_samples_split' : [2,3,4,5,6]
}
```

```python
rscv = RandomizedSearchCV(pipe2,param_distributions=parameter,cv=5,scoring='r2')
```

```python
rscv.fit(X_train,Y_train)
```

```python
rscv.best_score_
```

```python
rscv.best_params_
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
step2 = RandomForestRegressor()
```

```python
#creating the pipeline:
pipe3 = Pipeline([
    ('step1',step1),
    ('step2',step2),
])
```

```python
#training the data with our pipeline:
pipe3.fit(X_train,Y_train)
```
