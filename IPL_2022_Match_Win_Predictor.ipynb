```python
#importing deliveries data:
delivery = pd.read_csv('/content/IPL_Ball_by_Ball_2008_2022.csv',error_bad_lines=False)
delivery.head()
```

```python
# calculating total runs after each inning:
temp_df = delivery.groupby(['ID','innings'])['total_run'].sum().reset_index()
```

```python
temp_df
```

```python
# adding +1 for target
temp_df['target'] = temp_df['total_run'] + 1
```

```python
#we only need 1st inning score for further calculations to predict the scores for 2nd inning:
temp_df = temp_df[temp_df['innings'] == 1]
```

```python
#merging temp_df with matches df:
match_df = matches.merge(temp_df,on='ID')
```

```python
match_df.head()
```

```python
#working on teams:
match_df['Team1'].unique()
```

```python
#renaming same teams:

#'Deccan Chargers' --> 'Sunrisers Hyderabad'
match_df['Team1'] = match_df['Team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['Team2'] = match_df['Team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

#'Delhi Daredevils' --> 'Delhi Capitals'
match_df['Team1'] = match_df['Team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['Team2'] = match_df['Team2'].str.replace('Delhi Daredevils','Delhi Capitals')

#'Gujarat Lions' --> 'Gujarat Titans'
match_df['Team1'] = match_df['Team1'].str.replace('Gujarat Lions','Gujarat Titans')
match_df['Team2'] = match_df['Team2'].str.replace('Gujarat Lions','Gujarat Titans')

#'Kings XI Punjab' --> 'Punjab Kings'
match_df['Team1'] = match_df['Team1'].str.replace('Kings XI Punjab','Punjab Kings')
match_df['Team2'] = match_df['Team2'].str.replace('Kings XI Punjab','Punjab Kings')

#'Rising Pune Supergiants' --> 'Rising Pune Supergiant'
match_df['Team1'] = match_df['Team1'].str.replace('Rising Pune Supergiants','Rising Pune Supergiant')
match_df['Team2'] = match_df['Team2'].str.replace('Rising Pune Supergiants','Rising Pune Supergiant')

#'Pune Warriors'--> 'Rising Pune Supergiant'
match_df['Team1'] = match_df['Team1'].str.replace('Pune Warriors','Rising Pune Supergiant')
match_df['Team2'] = match_df['Team2'].str.replace('Pune Warriors','Rising Pune Supergiant')
```

```python
match_df['Team1'].unique()
```

```python
# keeping required cols:
match_df = match_df[['ID','City','Team1','Team2','WinningTeam','WonBy','total_run','target']]
```

```python
match_df.head()
```

```python
#merging match_df with delivery to get ball by ball details:
delivery_df = delivery.merge(match_df,on='ID')
```

```python
#keeping values where inning = 2 for training the model:
delivery_df = delivery_df[delivery_df['innings'] == 2]
```

```python
delivery_df.head()
```

```python
#calculating the current score after every ball:
delivery_df['current_score'] = delivery_df.groupby('ID')['total_run_x'].cumsum()
```

```python
#calculating runs left after every ball:
delivery_df['runs_left'] = delivery_df['target'] - delivery_df['current_score']
```

```python
#calculating balls left after every ball:
delivery_df['balls_left'] = 120 - (delivery_df['overs']*6 + delivery_df['ballnumber'])
```

```python
#calculating wickets left after every ball:
delivery_df['wickets_left'] = 10 - (delivery_df.groupby('ID')['isWicketDelivery'].cumsum())
```

```python
delivery_df
```

```python
#calculating current run rate(crr):
delivery_df['crr'] = (delivery_df['current_score']*6)/(120-delivery_df['balls_left'])
```

```python
#calculating required run rate(rrr):
delivery_df['rrr'] = (delivery_df['runs_left']*6)/(delivery_df['balls_left'])
```

```python
#defining a function to get results as 0 or 1
def result(row):
    return 1 if row['BattingTeam'] == row['WinningTeam'] else 0
```

```python
#applyng the result function:
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
delivery_df['result'].value_counts()
```

```python
delivery_df.info()
```

```python
#creating bowling team as there is no bowling team column in the df:
```

```python
def bowling_team(row):
    return row['Team1'] if row['Team2'] == row['BattingTeam'] else row['Team2']
```

```python
delivery_df['BowlingTeam'] = delivery_df.apply(bowling_team,axis=1)
```

```python
#checking the bowling column with other cols:
delivery_df[['Team1','Team2','BattingTeam','BowlingTeam']]
```

```python
# gathering the required cols for final df:
final_df = delivery_df[['BattingTeam', 'BowlingTeam', 'City', 'runs_left', 'balls_left', 'wickets_left', 'crr', 'rrr', 'target', 'result']]
```

```python
#shuffling the df as we have sorted data:
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df = final_df[final_df['balls_left'] > 0 ]
```

## Building the models:

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_df.iloc[:,:-1]
Y = final_df.iloc[:,-1]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train.shape
```

```python
Y_train.shape
```

## Creating pipelines to tranform cols and train the model:

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
# encoding batting team,bowling team and city:
step1 = ColumnTransformer(transformers=[
    ('ohe',OneHotEncoder(sparse=False,drop='first'),[0,1,2])
],remainder = 'passthrough')
```

## Logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
#assigning our model to step 2 :
step2 = LogisticRegression(C=1,solver='liblinear',max_iter=20,penalty='l1')
```

```python
#creating the pipeline:
pipe1 = Pipeline([
    ('step1',step1),
    ('step2',step2),
])
```

```python
#training the data with our pipeline:
pipe1.fit(X_train,Y_train)
```

```python
#testing the pipeline:
Y_pred1 = pipe1.predict(X_test)
```

```python
# evaluating the model:
from sklearn.model_selection import cross_val_score
```

```python
acc= cross_val_score(pipe1,X_train,Y_train,cv=5,scoring='accuracy').mean()
acc
```
